```python
import numpy as np
import cv2
import os
```

Train Bayou

```python
image_folder = '/content/drive/MyDrive/Assignment_CS669/Group01_Assignment2_code/SceneData/train/bayou'
output_file = '/content/drive/MyDrive/Assignment_CS669/Group01_Assignment2_code/SceneData/ttnpy/trainbayou.npy'
```

```python
image_data = []
for filename in os.listdir(image_folder):
    if filename.endswith(('.jpg')):
        image_path = os.path.join(image_folder, filename)
        img = cv2.imread(image_path)
        if img is not None:
            image_data.append(img)
```

```python
image_array = np.array(image_data)
```

```python
np.save(output_file, image_array, allow_pickle=True)
```
