## **Dataset:2**

```python
import os
import numpy as np

train_folder = "/content/drive/MyDrive/PR_Assignment-2/CS699-AS1-2-3-4-5/Group01_Assignment5_code/BOVW/train"

train_data = []
train_labels = []

class_folders = os.listdir(train_folder)
for i, class_folder in enumerate(class_folders):
    class_path = os.path.join(train_folder, class_folder)
    files = os.listdir(class_path)

    for file in files:
        file_path = os.path.join(class_path, file)
        bovw_representation = np.load(file_path)
        train_data.append(bovw_representation)
        train_labels.append(i)

X_train = np.vstack(train_data)
y_train = np.array(train_labels)
```

```python
test_folder = "/content/drive/MyDrive/PR_Assignment-2/CS699-AS1-2-3-4-5/Group01_Assignment5_code/BOVW/test"

test_data = []
test_labels = []

class_folders = os.listdir(test_folder)
for i, class_folder in enumerate(class_folders):
    class_path = os.path.join(test_folder, class_folder)
    files = os.listdir(class_path)

    for file in files:
        file_path = os.path.join(class_path, file)
        bovw_representation = np.load(file_path)
        test_data.append(bovw_representation)
        test_labels.append(i)
X_test = np.vstack(test_data)
y_test = np.array(test_labels)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.mixture import GaussianMixture
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix

# Define the number of mixtures for GMM
num_mixtures = [1, 2, 4, 8]

# Define different values of l (dimensions in PCA)
dimensions_l = [2, 4, 6, 8, 10, 12, 14]

for l in dimensions_l:
    # Apply PCA to reduce the dimensionality
    pca = PCA(n_components=l)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)

    for mixture in num_mixtures:
        # Train Gaussian Mixture Model
        gmm = GaussianMixture(n_components=mixture, random_state=42)
        gmm.fit(X_train_pca)

        # Predict on the test set
        y_pred = gmm.predict(X_test_pca)

        # Evaluate accuracy
        accuracy = accuracy_score(y_test, y_pred)

        precision = precision_score(y_test, y_pred, average='weighted')
        recall = recall_score(y_test, y_pred, average='weighted')
        conf_matrix = confusion_matrix(y_test, y_pred)

        print(f"Dimensions (l={l}), Mixtures={mixture}:")
        print(f"  Accuracy: {accuracy:.2f}")
        print(f"  Precision: {precision:.2f}")
        print(f"  Recall: {recall:.2f}")
        print("  Confusion Matrix:")
        print(conf_matrix)

        #print(f"Dimensions (l={l}), Mixtures={mixture}, Accuracy: {accuracy:.2f}")

        # Visualize the results (for 2D PCA)
        #if l == 4:
        plt.figure(figsize=(8, 6))
        plt.scatter(X_test_pca[:, 0], X_test_pca[:, 1], c=y_pred, cmap='viridis')
        plt.title(f'PCA (l={l}) with GMM Mixtures={mixture}')
        plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.mixture import GaussianMixture
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix

# Assuming you have loaded your data into X_train, X_test, y_train, y_test

# Define the number of mixtures for GMM
num_mixtures = [1, 2, 4, 8]

# Define different values of l (dimensions in PCA)
dimensions_l = [2, 4, 6, 8, 10, 12, 14]

for l in dimensions_l:
    # Apply PCA to reduce the dimensionality
    pca = PCA(n_components=l)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)

    # For original data
    gmm_original = GaussianMixture(n_components=4, random_state=42)
    gmm_original.fit(X_train)

    y_pred_original = gmm_original.predict(X_test)

    accuracy_original = accuracy_score(y_test, y_pred_original)
    precision_original = precision_score(y_test, y_pred_original, average='weighted')
    recall_original = recall_score(y_test, y_pred_original, average='weighted')
    conf_matrix_original = confusion_matrix(y_test, y_pred_original)

    print(f"For Original Data (l={l}):")
    print(f"  Accuracy: {accuracy_original:.2f}")
    print(f"  Precision: {precision_original:.2f}")
    print(f"  Recall: {recall_original:.2f}")
    print("  Confusion Matrix:")
    print(conf_matrix_original)

    # For PCA-reduced data
    for mixture in num_mixtures:
        # Train Gaussian Mixture Model
        gmm_pca = GaussianMixture(n_components=mixture, random_state=42)
        gmm_pca.fit(X_train_pca)

        # Predict on the test set
        y_pred_pca = gmm_pca.predict(X_test_pca)

        # Evaluate accuracy
        accuracy_pca = accuracy_score(y_test, y_pred_pca)

        # Calculate precision, recall, and confusion matrix for PCA-reduced data
        precision_pca = precision_score(y_test, y_pred_pca, average='weighted')
        recall_pca = recall_score(y_test, y_pred_pca, average='weighted')
        conf_matrix_pca = confusion_matrix(y_test, y_pred_pca)

        print(f"For PCA-reduced Data (l={l}, Mixtures={mixture}):")
        print(f"  Accuracy: {accuracy_pca:.2f}")
        print(f"  Precision: {precision_pca:.2f}")
        print(f"  Recall: {recall_pca:.2f}")
        print("  Confusion Matrix:")
        print(conf_matrix_pca)
```

```python
# Combine training and test data for PCA
X_combined = np.vstack([X_train, X_test])
y_combined = np.concatenate([y_train, y_test])

# Calculate eigenvalues during PCA
pca_combined = PCA()
pca_combined.fit(X_combined)

# Plot eigenvalues for both training and test data combined
plt.plot(np.arange(1, len(pca_combined.explained_variance_) + 1), np.sort(pca_combined.explained_variance_)[::-1])
plt.title('Eigenvalues in Descending Order by Principal Components (Combined Data)')
plt.xlabel('Principal Components')
plt.ylabel('Eigenvalues')
plt.grid(True)
plt.show()
```

```python
# Apply PCA for 2-dimensional reduced representations
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Plot 2D representations for training data
plt.figure(figsize=(10, 6))
for i in np.unique(y_train):
    plt.scatter(X_train_pca[y_train == i, 0], X_train_pca[y_train == i, 1], label=f'Class {i}')

plt.title('2D PCA for Training Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()

# Plot 2D representations for test data
plt.figure(figsize=(10, 6))
for i in np.unique(y_test):
    plt.scatter(X_test_pca[y_test == i, 0], X_test_pca[y_test == i, 1], label=f'Class {i}')

plt.title('2D PCA for Test Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```
